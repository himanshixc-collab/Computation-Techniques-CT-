```python
#Python program to compute accuracy, precision, recall, F1-score for a confusion matrix

#Confusion Matrix values
TP = 250
TN = 256
FP = 0
FN = 5

#Calculations
Accuracy = (TP + TN)/(TP+TN+FP+FN)
Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
F1_Score = 2*(Precision*Recall)/(Precision+Recall)

#Outputs
print("Accuracy: ", Accuracy)
print("Precision: ", Precision)
print("Recall: ", Recall)
print("F1 - Score : ", F1_Score)
```
